```python
titanic_data_copy.isna().sum() #checks the number (sum) of all the NaN values (isna) in the data 
```

```python
titanic_data_copy["Age"].max() #checks the maximum age in the data
```

```python
titanic_data_copy["Age"].min() #checks the minimum age in the data
```

```python
titanic_data_copy["Age"].fillna(titanic_data_copy["Age"].median(), inplace = True) #will fillna (fill NaN) values in the Age column with the median age of all the ages in the Age column. inplace = True 'authorises' the fillna command
```

```python
titanic_data_copy.isna().sum() #checks the number (sum) of all the NaN values (isna) in the data 
```

```python
titanic_data_copy["Embarked"].mode() #checks the mode (most common) value in the Embarked column
```

```python
titanic_data_copy["Embarked"].fillna(titanic_data_copy["Embarked"].mode()[0], inplace = True) #will fillna (fill NaN) values in the Embarked column with the mode value
```

```python
titanic_data_copy.isna().sum() #checks the number (sum) of all the NaN values (isna) in the data - now all changed in the data
```

# One-hot encoding

```python
dummies = []
cols = ["Pclass", "Sex", "Embarked"]
```

```python
for col in cols:
  dummies.append(pd.get_dummies(titanic_data_copy[col]))
```

```python
all_dummies = pd.concat(dummies, axis=1)
all_dummies
```

```python
titanic_data_copy = pd.concat((titanic_data_copy, all_dummies), axis=1)
titanic_data_copy = titanic_data_copy.drop(["Pclass", "Sex", "Embarked"], axis=1)
```
